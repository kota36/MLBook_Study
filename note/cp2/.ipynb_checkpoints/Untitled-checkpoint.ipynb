```python
import matplotlib.pylab as plt
import numpy as np
import pandas as pd
```

```python
class housePriceData:
    """　物件価格を読み込み、指定した建物の投球の散布図をプロットするクラス
    """
    
    def __init__(self, path):
        """pathのcsvファイルを読み込み、self.dataに格納する
        
        Args:
            path (str): csvファイルのパス
        """
        self.data = pd.read_csv(path) # dataframeの読み込み
        
    def plotScatter(self, titles, levels):
        """ 建物の等級を限定した散布図をプロットする
        
        Args:
            titles (list[str]): タイトル
            leves (list[int]): 建物の等級
        """
        # グラフの初期化
        fig = plt.figure()
        
        # 軸の範囲計算
        xrange = [np.min(self.data['GrLivArea'].values), np.max(self.data['GrLivArea'].values)]
        yrange = [np.min(self.data['SalePrice'].values), np.max(self.data['SalePrice'].values)]
        
        # 列数の計算
        ncol = int(len(titles)/2)
        
        for ind in np.arange(len(titles)):
            # グラフ位置の指定
            ax = fig.add_subplot(2, ncol, ind + 1)
            # タイトルの設定
            ax.set_title(titles[ind])
            # 散布図のプロット
            ax.plot(self.data[self.data['MSSubClass']==levels[ind]]['GrLivArea'].values, self.data[self.data['MSSubClass']==levels[ind]]['SalePrice'].values, '.')
            
            # 各軸の範囲とラベルの設定
            ax.set_xlim([xrange[0], xrange[1]])
            ax.set_ylim([yrange[0], yrange[1]])
            ax.set_xlabel('GrLivArea', fontsize=14)
            ax.set_ylabel('SalePrice', fontsize=14)
        
        # グラフ間に隙間を開ける
        plt.tight_layout()
        
        # グラフを表示する
        plt.show()
    
    
```

```python
# housePriceDataクラスのインスタンス化
myData = housePriceData('../..//data/house-price-advanced-regression-techniques/housetrain.csv')

# MSSubClassとタイトルのリストを作成
levels = [20, 30, 60, 70]

titles = []
titles.append('1-Story 1946 & Newer')
titles.append('2-Story 1945 & Older')
titles.append('1-Story 1946 & Newer')
titles.append('2-Story 1945 & Older')

# 散布図をプロット
myData.plotScatter(titles, levels)
```
