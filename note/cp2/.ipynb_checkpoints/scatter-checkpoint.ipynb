```python
import pandas as pd
import matplotlib.pylab as plt
```

### データの読み込み

```python
data = pd.read_csv('../../data/house-price-advanced-regression-techniques/housetrain.csv')
```

```python
data.head()
```

```python
data.describe()
```

#### 一階建て1945年以前建設（MSSubClass=30）の価格

```python
data[data['MSSubClass']==30]['SalePrice'].describe()
```

## 散布図のプロット

```python
fig = plt.figure()

# 横軸GrLivArea 縦軸SalePrice
ax = fig.add_subplot(1,2,1)  # グラフの位置指定（1行2列の1列目）
ax.plot(data['GrLivArea'], data['SalePrice'], '.')
ax.set_xlabel('GrLivArea', fontsize=14)  
ax.set_ylabel('SalePrice',fontsize=14)  

# 横軸MSSubClass 縦軸SalePrice
ax = fig.add_subplot(1, 2, 2)
ax.plot(data['MSSubClass'], data['SalePrice'], ',')
ax.set_xlabel('MSSubClass', fontsize=14)
ax.set_ylabel('SalePrice', fontsize=14)

fig.tight_layout()
plt.show()
```
